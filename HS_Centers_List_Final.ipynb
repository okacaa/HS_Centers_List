```python
import os
import numpy as np
import pandas as pd
```

```python
locs = pd.read_csv("https://eclkc.ohs.acf.hhs.gov/sites/default/files/locatordata/ALL_all.csv")
```

```python
z = pd.read_csv('zip_code_database.csv',
               dtype=object)
```

```python
z_ok = z.loc[z['state'] == 'OK', ['zip', 'state', 'county']]
```

```python
locs_ok = locs.loc[locs['state'] == 'OK', ['name', 'grantee', 'centerType', 'phone', 
                                           'address', 'city', 'state', 'zip']]
```

```python
locs_ok.info()
```

```python
y = locs_ok.zip
```

```python
y
```

```python
clean = y.str.split('-', expand=True)
```

```python
clean
```

```python
x = locs_ok.join(clean,
            how='left')
```

```python
x
```

```python
x.info()
```

```python
z_ok.columns = [0, 'state', 'county']
```

```python
fin = pd.merge(x,
               z_ok,
               how='left',
               on=0)
```
